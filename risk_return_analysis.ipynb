### Import the Data

```python
# Import the required libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Import the data by reading in the CSV file and setting the DatetimeIndex 
whales_df = pd.read_csv(Path('Resources/whale_navs.csv'), index_col="date", parse_dates=True, infer_datetime_format=True)

# Review the first 5 rows of the DataFrame
whales_df.head()
```

```python
# Prepare for the analysis by converting the dataframe of NAVs and prices to daily returns
# Drop any rows with all missing values
daily_returns = whales_df.pct_change().dropna()
# Review the first five rows of the daily returns DataFrame.
daily_returns.head()
```

### Analyze the Risk

Evaluate the risk profile of each portfolio by using the standard deviation and the beta.

```python
# Calculate and sort the standard deviation for all 4 portfolios and the S&P 500
# Review the standard deviations sorted smallest to largest
standard_deviation = daily_returns.std().sort_values()
standard_deviation
```

```python
# Calculate and sort the annualized standard deviation (252 trading days) of the 4 portfolios and the S&P 500 (daily STD * Trading Days ^1/2)
trading_days = 252
annual_standard_deviation = standard_deviation * trading_days ** (1/2)

# Review the annual standard deviations smallest to largest
annual_standard_deviation.sort_values()
```

```python
# Using the daily returns DataFrame and a 21-day rolling window, 
# plot the rolling standard deviation of the 4 portfolios and the S&P 500
# Include a title parameter and adjust the figure size
daily_returns.rolling(window=21).std().plot(kind='line', figsize=(14,7), title="21 Day Rolling Standard Deviations for Portfolios and S&P 500", legend=True)
```

```python
# Using the daily return data and a 21-day rolling window, plot the rolling standard deviation of just the 4 portfolios. 
# Include a title parameter and adjust the figure size
portfolios_dreturn_df.rolling(window=21).std().plot(kind='line', figsize=(14,7), title="21 Day Rolling Standard Deviations for Portfolios", legend=True)
```

### Analyze the Risk-Return Profile


```python
# Calculate the annual average return data for the for fund portfolios and the S&P 500
# Use 252 as the number of trading days in the year
annual_average_return = daily_returns.mean()* trading_days

# Review the annual average returns sorted from lowest to highest
annual_average_return.sort_values()
```

```python
# Calculate the annualized Sharpe Ratios for each of the 4 portfolios and the S&P 500.
annual_sharpe_ratio = annual_average_return/annual_standard_deviation

# Review the Sharpe ratios sorted lowest to highest
annual_sharpe_ratio.sort_values()
```

```python
# Visualize the Sharpe ratios as a bar chart
# Include a title parameter and adjust the figure size
annual_sharpe_ratio.plot(kind='bar', figsize=(14,7), title='Sharpe Ratios for Porfolios and S&P 500')
```

### Diversify the Portfolio

Based on previous analysis Berkshire Hathaway Inc and Tiger Global Management LLC are the two portfolios to further evaluate.

```python
# Calculate the variance of the S&P 500 using a rolling 60-day window.
sp500_variance_60_day = daily_returns['S&P 500'].rolling(window=60).var()

sp500_variance_60_day.tail()
```

##### Portfolio 1 - Berkshire Hathaway Inc.

```python
# Calculate the covariance using a 60-day rolling window 
# Review the last five rows of the covariance data
berkshire_covariance_60_day = daily_returns['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(daily_returns['S&P 500'])
berkshire_covariance_60_day.tail()
```

```python
# Calculate the beta based on the 60-day rolling covariance compared to the market (S&P 500)
berkshire_beta_60_day = berkshire_covariance_60_day/sp500_variance_60_day
# Review the last five rows of the beta information
berkshire_beta_60_day.tail()
```

```python
# Calculate the average of the 60-day rolling beta
avg_berkshire_beta_60_day = berkshire_beta_60_day.mean()
avg_berkshire_beta_60_day
```

```python
# Plot the rolling beta 
# Include a title parameter and adjust the figure size
berkshire_beta_60_day.plot(kind='line', figsize=(14,7), title='60-Day Rolling Beta for Berkshire Hathaway INC')
```

##### Portfolio 2 - Tiger Global Management LLC

```python
# Calculate the covariance using a 60-day rolling window 
tiger_covariance_60_day = daily_returns['TIGER GLOBAL MANAGEMENT LLC'].rolling(window=60).cov(daily_returns['S&P 500'])

# Review the last five rows of the covariance data
tiger_covariance_60_day.tail()
```

```python
# Calculate the beta based on the 60-day rolling covariance compared to the market (S&P 500)
tiger_beta_60_day = tiger_covariance_60_day/sp500_variance_60_day

# Review the last five rows of the beta information
tiger_beta_60_day.tail()
```

```python
# Calculate the average of the 60-day rolling beta
avg_tiger_beta_60_day = tiger_beta_60_day.mean()
avg_tiger_beta_60_day
```

```python
# Plot the rolling beta 
# Include a title parameter and adjust the figure size
tiger_beta_60_day.plot(kind='line', figsize=(14,7), title='60-Day Rolling Beta for Tiger Global Management LLC')
```
